```python
FEAT_TIER_MAP = {
'LEFT_PALM/BACK_LABEL': 'LEFT_ORIENTATION_PALM/BACK',
'LEFT_TIPS/WRIST_LABEL': 'LEFT_ORIENTATION_TIPS/WRIST',
'LEFT_RADIAL/URNAL_LABEL': 'LEFT_ORIENTATION_RADIAL/URNAL',
'RIGHT_PALM/BACK_LABEL': 'RIGHT_ORIENTATION_PALM/BACK', 
'RIGHT_TIPS/WRIST_LABEL': 'RIGHT_ORIENTATION_TIPS/WRIST', 
'RIGHT_RADIAL/URNAL_LABEL': 'RIGHT_ORIENTATION_RADIAL/URNAL', 
'IN_FRONT_LABEL': 'MANNER_SPACIAL_IN_FRONT_OF',
'ABOVE/BELOW_LABEL': 'MANNER_SPACIAL_ABOVE', 
'LEFT_SETTING_LABEL': 'LEFT_LOCATION_SETTING', 
'RIGHT_SETTING_LABEL': 'RIGHT_LOCATION_SETTING',
'LEFT_REPETITION_LABEL': 'MANNER_SPACIAL_REPETITION', 
}
```

```python
HAND = 'RIGHT'
for col, tier_name in FEAT_TIER_MAP.items():
    seg_start = 0
    seg_end = 0
    seg_label = None
    
    record_rate = 0 # How many frames that will be extracted per quarter of a second    
    
    for idx, label in list(enumerate(df[col].replace(pd.NA, None).to_list())):
        idx = int((idx / fps) *1000)
        if seg_label != label:
            if seg_label != None:
                try:
                    #print(seg_label, seg_start, idx-1)
                    elan.add_annotation(tier_name, seg_start, idx-1, value=seg_label, svg_ref=None)
                except ValueError: 
                    #print('Cannot write annotation:')
                    #print(seg_label, seg_start, idx-1)
                    #print(df[col][seg_start-5: idx+5])
                    #print('*'*20)
                    pass
                    
            seg_start = idx
            seg_label = label
        
```
