```python
def print_tree(tier_name, indent='|-'):
    children = elan.get_child_tiers_for(tier_name)
    if children:
        for tt in children:
            print(indent, tt)
            print_tree(tt, indent='|----'+indent)

print('MANUAL')
print_tree('MANUAL')
```
